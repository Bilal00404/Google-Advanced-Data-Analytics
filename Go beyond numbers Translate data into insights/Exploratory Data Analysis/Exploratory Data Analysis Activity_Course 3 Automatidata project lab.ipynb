### Boxplots

Perform a check for outliers on relevant columns such as trip distance and trip duration. Remember, some of the best ways to identify the presence of outliers in data are box plots and histograms. 

**Note:** Remember to convert your date columns to datetime in order to derive total trip duration.  

```python
# Convert data columns to datetime
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

**trip distance**

```python
# Create box plot of trip_distance

#Set figure style
sns.set_style("darkgrid")

# Set figure size
plt.figure(figsize=(9,3))

# Plot figure
sns.boxplot(x=df['trip_distance'])

# Add labels and title
plt.title('Boxplot of trip distance')

# Show the plot
plt.show()
```

```python
# Create histogram of trip_distance

# Create a histogram
sns.histplot(df['trip_distance'], bins=range(0,27,1))

# Add labels and title
plt.xlabel('Trip Distance in Miles')
plt.ylabel('Frequency')
plt.title('Frequency distribution of trip distance')

# Show the plot
plt.show()
```

* **The distribution of trip distance is right skewed. There are fewer data points as distance increases. The majority of distance travelled is below 2 miles.** 

**total amount**

```python
# Create box plot of total_amount

plt.figure(figsize=(9,3))
sns.boxplot(x=df['total_amount'])

# Add labels and title
plt.title('Boxplot of total amount')

# Show the plot
plt.show()
```

```python
# Create histogram of total_amount
plt.figure(figsize=(9,5))
ax = sns.histplot(df['total_amount'], bins=range(-5,102,5))

# set ticks
ax.set_xticks(range(-5,102,5))

# Add labels and title
plt.xlabel('Total amount in dollars ($)')
plt.ylabel('Frequency')
plt.title('Distribution of total amount in dollars ($)')

# Show the plot
plt.show()
```

* **The distribution of total amount is right skewed. Most of the data points are between 5 and 15 dollars. The number of data points are fewer as the total amount increases.**

**tip amount**

```python
# Create box plot of tip_amount
plt.figure(figsize=(9,3))
sns.boxplot(x=df['tip_amount'])

# Add labels and title
plt.title('Boxplot of tip amount')

# Show the plot
plt.show()
```

```python
# Create histogram of tip_amount
plt.figure(figsize=(10,6))
ax = sns.histplot(df['tip_amount'], bins=range(0,22,1))

# set ticks
ax.set_xticks(range(0,22,1))

# Add labels and title
plt.xlabel('Tip amount in dollars($)')
plt.ylabel('Frequency')
plt.title('Distribution of tip amount')

# Show the plot
plt.show()
```

* **The distribution of total amount is right skewed. The total number of points decrease as the tip amount increase. Most of the data points are between 0 and 3 dollars.**

**tip_amount by vendor**

```python
# Create histogram of tip_amount by vendor

plt.figure(figsize=(10,5))
ax = sns.histplot(data=df, x='tip_amount', 
                  hue='VendorID', 
                  bins=range(0,22,1), 
                  multiple='stack',
                  palette='pastel')

# Set ticks
ax.set_xticks(range(0,22,1))

# Add labels and title
plt.xlabel('Tip amount in dollars($)')
plt.ylabel('Frequency')
plt.title('Distribution of tip amount by VendorID')

# Show the plot
plt.show()
```

* **The distribution of tip amount by both vendors is right skewed. Although it can be seen that the tip amounts for vendor ID=1 are greater in number than tip amounts for vendor ID=2.**

Next, zoom in on the upper end of the range of tips to check whether vendor one gets noticeably more of the most generous tips.

```python
# Create histogram of tip_amount by vendor for tips > $10 
tip_greater_10 = df[df['tip_amount']>10]

plt.figure(figsize=(10,5))
ax = sns.histplot(data=tip_greater_10, x='tip_amount', 
                  hue='VendorID', 
                  bins=range(10,25,1), 
                  multiple='stack',
                  palette='pastel')


#Set ticks
ax.set_xticks(range(10,25,1))


# Add labels and title
plt.xlabel('Tip amount in dollars($)')
plt.ylabel('Frequency')
plt.title('Distribution of tip amount (greater than 10$) by VendorID')

# Show the plot
plt.show()
```

* **The proportion of tip amounts are maintained even at these higher tip amounts.**

**Mean tips by passenger count**

Examine the unique values in the `passenger_count` column.

```python
df.value_counts('passenger_count')
```

* **Most of rides were with only single passenger. There were almost 700 rides where the passenger count was 6. There were 33 rides where the passenger count was zero, these points should be removed from the analysis unless there is an explanation for these points.**

```python
# Calculate mean tips by passenger_count
pasngr_avg_tip = df.groupby('passenger_count').agg({'tip_amount':'mean'})
pasngr_avg_tip = pasngr_avg_tip.sort_values('tip_amount', ascending=False).reset_index()
pasngr_avg_tip
```

```python
# Create bar plot for mean tips by passenger count

# Remove points where the passenger count is zero
data = pasngr_avg_tip[1:]

# Plot the figure
plt.figure(figsize=(9,5))
ax = sns.barplot(data=data, x='passenger_count', y='tip_amount')

# Add labels and title
plt.xlabel('Passenger Count')
plt.ylabel('Mean tip amount in dollars($)')
plt.title('Mean tips by passenger count')

# Show the plot
plt.show()
```

* **Mean tip amount does not vary much with passenger count.**

**Create month and day columns**

```python
# Create a month column
df['month'] = df['tpep_pickup_datetime'].dt.strftime('%B').str[0:3]

# Create a day column
df['day_of_week'] = df['tpep_pickup_datetime'].dt.strftime('%A').str[0:3]
```

**Plot total ride count by month**

Begin by calculating total ride count by month.

```python
# Get total number of rides for each month
monthly_rides = df.value_counts('month')
monthly_rides
```

Reorder the results to put the months in calendar order.

```python
# Reorder the monthly ride list so months go in order
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

monthly_rides = monthly_rides.reindex(index=months)
monthly_rides
```

```python
# Show the index
monthly_rides.index
```

```python
# Create a bar plot of total rides per month

plt.figure(figsize=(10,5))
sns.barplot(x=monthly_rides.index, y=monthly_rides)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Total rides')
plt.title('Total rides per month')

# Show the plot
plt.show()
```

* **Total rides by month are consistent. Although July, August, September and February have slightly lower number of rides compared to other months.**

**Plot total ride count by day**

Repeat the above process, but now calculate the total rides by day of the week.

```python
# Repeat the above process, this time for rides by day

weekly_rides = df.value_counts('day_of_week')

week_days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
weekly_rides = weekly_rides.reindex(index=week_days)
weekly_rides
```

```python
# Create bar plot for ride count by day
plt.figure(figsize=(10,5))
ax = sns.barplot(x=weekly_rides.index, y=weekly_rides)

# Add labels and title
plt.xlabel('Day of the week')
plt.ylabel('Total rides')
plt.title('Total rides per day of the week')

# Show the plot
plt.show()
```

* **The total number of rides is consistent from wednesday through saturday. Although the total number of rides are fewer from sunday to tuesday than other days of the week. Surprisingly monday and sunday have low number of rides.**

**Plot total revenue by day of the week**

Repeat the above process, but now calculate the total revenue by day of the week.

```python
# Repeat the process, this time for total revenue by day

revenue_weekday = df.groupby('day_of_week').agg({'total_amount':'sum'})
revenue_weekday = revenue_weekday.reindex(index=week_days)
revenue_weekday
```

```python
# Create bar plot of total revenue by day

plt.figure(figsize=(10,5))
sns.barplot(x=revenue_weekday.index, y=revenue_weekday['total_amount'])

# Add labels and title
plt.xlabel('Day of the week')
plt.ylabel('Total revenue in dollars($)')
plt.title('Total revenue per day of the week')

# Show the plot
plt.show()
```

* **From the graph it is clear that the highest revenue generated on thursdays and lowest revenue generated on sundays and mondays.**

**Plot total revenue by month**

```python
# Repeat the process, this time for total revenue by month

revenue_month = df.groupby('month').agg({'total_amount':'sum'})
revenue_month = revenue_month.reindex(index=months)
revenue_month
```

```python
# Create a bar plot of total revenue by month

plt.figure(figsize=(10,5))
ax = sns.barplot(x=revenue_month.index, y=revenue_month['total_amount'])

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Total revenue in dollars($)')
plt.title('Total revenue per month')

# Show the plot
plt.show()
```

* **From the graph it is clear that the least revenue generated months are July, August, September and February. These were the months with fewer number of rides that might be the reasen for the less revenue of these months compared to other months.**

```python
# Remove Extreme outliers
data_scatter = df[(df['fare_amount']>0) & (df['fare_amount']<450)]

# Plot
plt.figure(figsize=(10,6))
sns.scatterplot(x=data_scatter['fare_amount'], y=data_scatter['trip_distance'])

# Add labels and title
plt.xlabel('Fare amount')
plt.ylabel('Trip distance')
plt.title('Trip distance and fare amounts')

# Show the plot
plt.show()
```

* **Notice there are points in the graph where the trip distance is zero but still have fare amounts. These points are outliers and it is better to remove these outliers from the data before building any model.**

**Plot mean trip distance by drop-off location**

```python
# Get number of unique drop-off location IDs
print('Nmber of unique drop-off location IDs: ', len(df.value_counts('DOLocationID')))
```

```python
# Calculate the mean trip distance for each drop-off location
mean_Dloc_dist = df.groupby('DOLocationID').agg({'trip_distance': 'mean'})

# Sort the results in descending order by mean trip distance
mean_Dloc_dist = mean_Dloc_dist.sort_values('trip_distance', ascending=True)
mean_Dloc_dist
```

```python
# Create a bar plot of mean trip distances by drop-off location in ascending order by distance

plt.figure(figsize=(12,6))
ax = sns.barplot(x=mean_Dloc_dist.index, y=mean_Dloc_dist['trip_distance'], 
                 order=mean_Dloc_dist.index)

ax.set_xticks([])
# Add labels and title
plt.xlabel('Drop-off location')
plt.ylabel('Mean trip distance')
plt.title('Mean trip distances by drop-off location')

# Show the plot
plt.show()
```

**Histogram of rides by drop-off location**

First, check to whether the drop-off locations IDs are consecutively numbered. For instance, does it go 1, 2, 3, 4..., or are some numbers missing (e.g., 1, 3, 4...). If numbers aren't all consecutive, the histogram will look like some locations have very few or no rides when in reality there's no bar because there's no location. 

```python
# Check if all drop-off locations are consecutively numbered
df['DOLocationID'].max() - len(set(df['DOLocationID']))
```

To eliminate the spaces in the historgram that these missing numbers would create, sort the unique drop-off location values, then convert them to strings. This will make the histplot function display all bars directly next to each other. 

```python
# DOLocationID column is numeric, so sort in ascending order

plt.figure(figsize=(16,6))
DO_locations = df['DOLocationID'].sort_values(ascending=True)

# Convert to string
DO_locations = DO_locations.astype('str')

# Plot
ax = sns.histplot(DO_locations)

plt.xticks([])
# Add labels and title
plt.xlabel('Drop-off locations')
plt.ylabel('Frequency')
plt.title('Histogram of rides by drop-off location')

# Show the plot
plt.show()
```

* **Notice the drop off locations where the count is more than 200. These places might be famous places.**
